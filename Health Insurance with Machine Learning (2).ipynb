```python
# Create the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='age', y='charges', hue='smoker', palette='Set1')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter Plot of Charges vs. Age')

# Adjust layout to fit everything
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Ensure all columns are numerical for correlation calculation
numerical_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
try:
    correlation_matrix = numerical_df.corr(method='pearson')  # Change 'pearson' to your preferred method
    print("Correlation matrix computed successfully.")
except Exception as e:
    print(f"Error computing correlation matrix: {e}")

# Display the correlation matrix
print(correlation_matrix)




```

```python
# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add a title
plt.title('Correlation Matrix Heatmap')

# Adjust layout to fit everything
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a copy of the DataFrame to numerize the categorical columns
df_numerized = df.copy()

# Initialize LabelEncoder
label_encoders = {}

# Convert all categorical columns to numerical values
for column in df_numerized.columns:
    if df_numerized[column].dtype == 'object':
        le = LabelEncoder()
        df_numerized[column] = le.fit_transform(df_numerized[column])
        label_encoders[column] = le

# Display the numerized DataFrame
print(df_numerized.head())
```

```python
# Compute the correlation matrix
correlation_matrix = df_numerized.corr()

# Display the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add a title
plt.title('Numerized Correlation Matrix Heatmap')

# Adjust layout to fit everything
plt.tight_layout()

# Show the plot
plt.show()
```
